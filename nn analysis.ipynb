```python
import pandas as pd
import matplotlib.pyplot as plt
import re
column_names = ['iteration', 'mse_train', 'mse_val', 'mse_test', 'acc_train', 'acc_val', 'acc_test', 'f1_train', 'f1_val', 'f1_test', 'time']
```

**BackPropagation**

```python
bb_spam = pd.read_csv('output/spam/backprop_NN_LOG.txt', header=None, names=column_names, index_col='iteration')
bb_spam['mse_train'].plot(title='bb mse_train')
bb_spam['mse_val'].plot(title='bb mse_val')
```

**RHC**

```python
rhc_spam = pd.read_csv('output/spam/rhc_NN_LOG.txt', header=None, names=column_names, index_col='iteration')
pd.DataFrame(dict(map(lambda f: (f"{f[0]}_mse_train", f[1]['mse_train']), list([('bb_spam',bb_spam), ('rhc_spam',rhc_spam)])))).plot(title='spam mse train')
pd.DataFrame(dict(map(lambda f: (f"{f[0]}_mse_val", f[1]['mse_val']), list([('bb_spam',bb_spam), ('rhc_spam',rhc_spam)])))).plot(title='spam mse val')
pd.DataFrame(dict(map(lambda f: (f"{f[0]}_time", f[1]['time']), list([('bb_spam',bb_spam), ('rhc_spam',rhc_spam)])))).plot(title='spam time')
```

**Simulated Anneleaing**

```python
sa_files = [
    'output/spam/SA0.05_NN_LOG.txt',
    'output/spam/SA0.35_NN_LOG.txt', 
    'output/spam/SA0.55_NN_LOG.txt',
    'output/spam/SA0.95_NN_LOG.txt',
    'output/spam/SA0.7_NN_LOG.txt'
]

def read_sa(file_name):
    return (re.search("output/spam/(SA[0-9\.]+)", file_name).group(1), pd.read_csv(file_name, header=None, names=column_names, index_col='iteration'))

sa_dfs = dict(list(map(read_sa, sa_files)))
pd.DataFrame(dict(map(lambda f: (f"{f[0]}_mse_train", f[1]['mse_train']), sa_dfs.items()))).plot(title="SA mse train")
pd.DataFrame(dict(map(lambda f: (f"{f[0]}_mse_val", f[1]['mse_val']), sa_dfs.items()))).plot(title="SA mse val")
pd.DataFrame(dict(map(lambda f: (f"{f[0]}_time", f[1]['time']), sa_dfs.items()))).plot(title="SA time")
```
